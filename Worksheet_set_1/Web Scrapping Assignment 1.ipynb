```python
# 1. Write a python program to display all the header tags from wikipedia.org and make data frame.

from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

def wiki(input):
    page =requests.get(input)  # Requesting data from site
    page

    soup =BeautifulSoup(page.content)   # Downloading content
    soup

    tag =[]

    str1 ='central-featured-lang lang'
    count =1

    while count <=9:
        count1 =str(count)
   
        for i in soup.find_all('div', class_=str1+count1):
            tag.append(i.text)
        
        count+=1

    final_tag=[]
    i=0
    for each in tag:
        a =re.split('\n+', tag[i],)
        final_tag.append(a[1])
        i+=1
    
    df =pd.DataFrame({'TAG': final_tag})
    return df
    
wiki('https://www.wikipedia.org/')

```

```python
''' 2. Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice) 
from https://presidentofindia.nic.in/former-presidents.htm and make data frame.'''

from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

def president(input):
    page =requests.get(input)
    page

    soup =BeautifulSoup(page.content)
    soup

    name =[]
    termofoffice =[]



    for i in soup.find_all('div', class_ ='desc-sec'):
        name.append(i.h3.text)
        termofoffice.append(i.h5.text)
        
    df =pd.DataFrame({'President Name': name, 'Term': termofoffice})
    return df

president('https://presidentofindia.nic.in/former-presidents')

```

```python
'''
5. Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
i) Headline
ii) Time
iii) News Link
'''

from bs4 import BeautifulSoup
import requests
import pandas as pd
    
def news_site(input):
    
    page =requests.get(input)
    page

    soup =BeautifulSoup(page.content)
    soup

    # Extracting Headlines
    headline =[]
    for i in soup.find_all('a', class_ ='LatestNews-headline'):
        headline.append(i.text)
        
    # Extracting Link
    link =[]
    for i in soup.find_all('a', class_ ='LatestNews-headline'):
        link.append(i.get('href'))
    

    # Extracting Time
    time =[]
    for i in soup.find_all('time', class_ ='LatestNews-timestamp'):
        time.append(i.text)

             
    # Converting details into dataframe
    df =pd.DataFrame({'Headline': headline, 'Time': time, 'URL': link})
    return df
news_site('https://www.cnbc.com/world/?region=world')

# Converting details into dataframe
```

```python
'''
6. Write a python program to scrape the details of most downloaded articles from AI in last 90 days.
https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details 
and make data frame-
i) Paper Title
ii) Authors
iii) Published Date
iv) Paper URL
'''
from bs4 import BeautifulSoup
import requests
import pandas as pd

def journals(input):
    

    page =requests.get(input)
    page

    soup =BeautifulSoup(page.content)
    soup

    # Extracting Title
    title =[]
    for i in soup.find_all('h2', class_ ='sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg'):
        title.append(i.text)


    # Extracting Author
    author =[]
    for i in soup.find_all('span', class_ ='sc-1w3fpd7-0 dnCnAO'):
        author.append(i.text)


    # Extracting Published Date
    pub_date =[]
    for i in soup.find_all('span', class_ ='sc-1thf9ly-2 dvggWt'):
        pub_date.append(i.text)


    # Extracting URL
    url =[]
    for i in soup.find_all('a', class_ ='sc-5smygv-0 fIXTHm'):
        url.append(i.get('href'))


    df =pd.DataFrame({'Paper Title': title, 'Authors': author, 'Published Date': pub_date, 'Paper URL': url})
    return df
    
journals('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
```
